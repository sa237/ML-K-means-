```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage import io
```

```
image = io.imread('ML_Image.jpg')
io.imshow(image)
io.show()
```

```
rows = image.shape[0]
cols = image.shape[1]
image = image/255
X = image.reshape(image.shape[0]*image.shape[1],3)
```

For k=2


```
# number of clusters
K = 2 
# number of iterations the k-mean has
max_iters = 50 
```

```
import random
def init_centroids(X,K):
    cent = random.sample(list(X),K)
    return cent
```

```
def closest_centroids(X,cent):
    K = np.size(cent,0)
    idx = np.zeros((np.size(X,0),1))
    arr = np.empty((np.size(X,0),1))
    for i in range(0,K):
        y = cent[i]
        temp = np.ones((np.size(X,0),1))*y
        b = np.power(np.subtract(X,temp),2)
        a = np.sum(b,axis = 1)
        a = np.asarray(a)
        a.resize((np.size(X,0),1))
        arr = np.append(arr, a, axis=1)
    arr = np.delete(arr,0,axis=1)
    idx = np.argmin(arr, axis=1)
    return idx
```

```
import numpy.matlib
def compute_centroids(X,idx,K):
    n = np.size(X,1)
    centroids = np.zeros((K,n))
    for i in range(0,K):
        c_i = idx==i
        c_i = c_i.astype(int)
        total_number = sum(c_i);
        c_i.resize((np.size(X,0),1))
        total_matrix = np.matlib.repmat(c_i,1,n)
        c_i = np.transpose(c_i)
        total = np.multiply(X,total_matrix)
        centroids[i] = (1/total_number)*np.sum(total,axis=0)
    return centroids
```

```
def run_kMean(X,initial_centroids,max_iters):
    m = np.size(X,0)
    n = np.size(X,1)
    K = np.size(initial_centroids,0)
    centroids = initial_centroids
    previous_centroids = centroids
    idx = np.zeros((m,1))
    for i in range(1,max_iters):
        idx = closest_centroids(X,centroids)
        centroids = compute_centroids(X,idx,K)
    return centroids,idx
```

```
initial_centroids = init_centroids(X,K)
centroids,idx = run_kMean(X,initial_centroids,max_iters)
print(np.shape(centroids))
print(np.shape(idx))
```

```
idx = closest_centroids(X,centroids)
X_recovered = centroids[idx]
print(np.shape(X_recovered))
X_recovered = np.reshape(X_recovered, (rows, cols, 3))
print(np.shape(X_recovered))
```

```
import imageio
imageio.imwrite('ML_Image.jpg', X_recovered)
image_compressed = io.imread('ML_Image.jpg')
io.imshow(image_compressed)
io.show()
```
